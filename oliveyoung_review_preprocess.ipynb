### 레이블 인코딩

```python
# 레이블 인코딩 매핑 딕셔너리 정의
label_mapping = {"아주 만족해요": 3, "보통이에요": 2, "다소 아쉬워요": 1, None: 0}

# "absorption" 열에 대한 레이블 인코딩
df["absorption_encoded"] = df["absorption"].map(label_mapping)
```

```python
# ['보통이에요', '자극없이 순해요', '자극이 느껴져요', nan]
label_mapping = {"자극없이 순해요": 3, "보통이에요": 2, "자극이 느껴져요": 1, None: 0}

# "absorption" 열에 대한 레이블 인코딩
df["stimulation_encoded"] = df["stimulation"].map(label_mapping)
```

```python
# ['부드러워요', '보통이에요', '거친 편이에요', nan]
label_mapping = {"부드러워요": 3, "보통이에요": 2, "거친 편이에요": 1, None: 0}

# "absorption" 열에 대한 레이블 인코딩
df["texture_encoded"] = df["texture"].map(label_mapping)
```
