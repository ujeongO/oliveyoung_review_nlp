```python
import csv
from tqdm import tqdm
```

```python
sanitary_type = "pad"
goods_id = "A000000172347"
page = 72
brand_name = "ì˜ˆì§€ë¯¸ì¸"
```

```python
with open("./olivey_review_pad.csv", "a", encoding="utf-8") as fa:
    writer = csv.writer(fa)
    # writer.writerow(["item_number","brand_name","sanitary_type","item_option", "review_content","absorption", "stimulation", "texture"])
    review_list = []
    for i in tqdm(range(page), desc="Fetching Reviews"):
        url = url_template.format(target_id=goods_id, page=i+1)
        response = requests.get(
            url=url,
            headers={"x-requested-with": "XMLHttpRequest"})
        data=response.json()
        review_list.append(data["gdasList"]) 
    for i in tqdm(range(len(review_list)), desc="Writing to csv"):
        for review in review_list[i]:
            review_option = review["itemNm"]
            item_num = review["goodsNo"]
            review_content = review["gdasCont"]
            eval_list = review.get('evalList')
            if eval_list is None:
                eval_list = []
            eval_dict = {}
            for eval_item in eval_list:
                eval_dict[eval_item['evalQstCont']] = eval_item['evalAnsCont']
            writer.writerow(
                    [item_num, brand_name, sanitary_type, review_option, review_content, 
                    eval_dict.get('í¡ìˆ˜ë ¥', ''), eval_dict.get('ìê·¹ë„', ''), eval_dict.get('ì´‰ê°', '')
                    ])
```
