```python
def tokenize(sentence):
    try:
        tokens = komoran.pos(sentence)
        nouns = [token for token, tag in tokens if tag in ("NNP", "NNG", "XR", "VA")]
        return nouns
        # return komoran.pos(sentence)
    except:
        # print("예외!", sentence)
        print("예외!")
        return []
```

```python
tokenize(sentence)
```

```python
sampled_df = df.sample(100, random_state=1234)
```

```python
from tqdm import tqdm

tqdm.pandas()
sampled_df["nouns"] = sampled_df["review_content"].progress_apply(lambda x: tokenize(x))
```
